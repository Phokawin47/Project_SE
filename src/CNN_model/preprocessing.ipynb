```python
import os
```

```python
data_path = "archive/Dataset"

paths = os.listdir(data_path)
data_list = []

for p in paths:
    digit = p.split(" ")[1]
    files = os.listdir(data_path + "/" + p)
    data_list.extend([("/".join([data_path, p, f]), digit) for f in files])

print(f"Total samples: {len(data_list)}")

with open("data_list.txt", "w") as f:
    for item in data_list:
        f.write(f"{item[0]},{item[1]}\n")
```
